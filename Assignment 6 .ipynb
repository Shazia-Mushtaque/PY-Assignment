```python
#Code for creating a class ‘Car’
class Car():
     def __init__(self,model,colour,make,name,fuel):
        #5 attributes of car
        self.model = model
        self.colour = colour
        self.make = make
        self.name = name
        self.fuel = fuel

Car1 = Car(2008,'diamond silver','C200','Mercedes Benz','diesel')

#print Car1

print("Details of Car1")
print("-*-*-*-*-*-*-*-*-")
print(Car1.name)
print(Car1.model)
print(Car1.colour)
print(Car1.make)
print(Car1.fuel)


Car2 = Car(2020,'BMW230','Black','BMW','diesel')

#print Car2

print('\n'"Details of Car2")
print("-*-*-*-*-*-*-*-*-")
print(Car2.name)
print(Car2.model)
print(Car2.colour)
print(Car2.make)
print(Car2.fuel)

Car3 = Car(2019,'Camry','White','Toyota','diesel')

#print Car3

print('\n'"Details of Car3")
print("-*-*-*-*-*-*-*-*-")
print(Car3.name)
print(Car3.model)
print(Car3.colour)
print(Car3.make)
print(Car3.fuel)

Car4 = Car(2019,'Highlander','Grey','Toyota','diesel')

#print Car4

print('\n'"Details of Car4")
print("-*-*-*-*-*-*-*-*-")
print(Car4.name)
print(Car4.model)
print(Car4.colour)
print(Car4.make)
print(Car4.fuel)

Car5 = Car(2019,'F8 Tributo','Red','Fierrari','diesel')

#print Car5

print('\n'"Details of Car5")
print("-*-*-*-*-*-*-*-*-")
print(Car5.name)
print(Car5.model)
print(Car5.colour)
print(Car5.make)
print(Car5.fuel)
```
